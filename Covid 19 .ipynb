```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_json('https://www.mohfw.gov.in/data/datanew.json')
```

```python
data
```

```python
data  = data.iloc[:-1,1:]
```

```python
data
```

```python
data.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

total_states = np.arange(len(data['state_name']))
```

```python
# array value we will use for plotting and after plot it we will replace it later

total_states
```

```python
data['active']
```

```python
max(data['positive'])
```

# Total Positive Cases Based On States In India 

```python
# we are seeing what kind of highest count and what state have lowest count of patients

from matplotlib.pyplot import figure
figure(num=None, figsize=(9,6), dpi=80,facecolor='w',edgecolor='k')
plt.barh(total_states, data['positive'],align='center',alpha=0.5,
        color=(1,0,0),
        edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states, data['state_name'])
plt.xlim(1,max(data['positive']+100))
plt.xlabel('Positive number of cases')
plt.title('Total Corona Virus Cases')
plt.show()
```

### Active Number of cases based on States

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9,6), dpi=80,facecolor='w',edgecolor='k')
plt.barh(total_states, data['active'],align='center',alpha=0.5,
        color=(0,1,0),
        edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states, data['state_name'])
plt.xlim(1,max(data['active']+100))
plt.xlabel('Active number of cases')
plt.title('Active Corona Virus Cases')
plt.show()
```

### Total Number of Death Cases  

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9,6), dpi=80,facecolor='w',edgecolor='k')
plt.barh(total_states, data['death'],align='center',alpha=0.5,
        color=(0,0,1),
        edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states, data['state_name'])
plt.xlim(1,max(data['death']+100))
plt.xlabel('Death number of cases')
plt.title('Death Corona Virus Cases')
plt.show()
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(9,6), dpi=80,facecolor='w',edgecolor='k')
plt.barh(total_states, data['new_active'],align='center',alpha=0.5,
        color=(0,0,0.2,0),
        edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states, data['state_name'])
plt.xlim(1,max(data['new_active']+100))
plt.xlabel('New-Active number of cases')
plt.title('New-Active Corona Virus Cases')
plt.show()
```
